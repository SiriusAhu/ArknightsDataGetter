```python
import pandas as pd
from utils.tools import change_column
import ast

DATA = "data/03-sorted.csv"

data = pd.read_csv(DATA)
data.head()
```

```python
# 只保留需要的列

remain_column = ["干员名称", "稀有度", "职业", "再部署时间", "部署费用", "阻挡数", "攻击间隔", "生命上限", "攻击", "防御", "法术抗性", "攻击范围"]
data_ = data[remain_column]
data_.head()
```

## 进一步处理数据

```python
# 将部署费用分为 部署费用min和部署费用max
# 部署费用：18-20 -> 部署费用min：18 部署费用max：20
data_['部署费用min'] = data_['部署费用'].apply(lambda x: int(x.split('→')[0]))
data_['部署费用max'] = data_['部署费用'].apply(lambda x: int(x.split('→')[-1]))
data_.drop('部署费用', axis=1, inplace=True)
data_.head()
```

```python
# 尝试取 生命上限、攻击、防御、法术抗性 的第4个值，如果不是数字，则尝试取第3个值...
# 如果都不是数字，则assert 找不到
# 例如：生命上限：['699', '958', '1198', '1480'] -> 1480
def get_max_value(x):
    values = ast.literal_eval(x)
    for i in range(3, -1, -1):
        try:
            return int(values[i])
        except ValueError:
            continue
    assert False, "找不到最大值"
data_['生命上限_normal'] = data_['生命上限'].apply(get_max_value)
data_['攻击_normal'] = data_['攻击'].apply(get_max_value)
data_['防御_normal'] = data_['防御'].apply(get_max_value)
data_['法术抗性_normal'] = data_['法术抗性'].apply(get_max_value)
# 倒数5个
data_.tail()
```

```python
# 将 生命上限、攻击、防御、法术抗性 第5个值单独取出来，作为额外的属性_trust
# 例如：生命上限：['699', '958', '1198', '1480', '0'] -> 0 作为 生命上限_trust
def get_trust_value(x):
    return x.split("'")[-2]
data_['生命上限_trust'] = data_['生命上限'].apply(get_trust_value)
data_['攻击_trust'] = data_['攻击'].apply(get_trust_value)
data_['防御_trust'] = data_['防御'].apply(get_trust_value)
data_['法术抗性_trust'] = data_['法术抗性'].apply(get_trust_value)
data_.head()
```

```python
# 将 生命上限、攻击、防御、法术抗性 4列删去
drop_list = ['生命上限', '攻击', '防御', '法术抗性']
data_ = data_.drop(columns=drop_list)
data_.tail()
```

```python
# 攻击范围只取最大值
# [9, 10, 10] -> 10
# [9, -, -] -> 9
def get_attack_range_value(x):
    """获取攻击范围的最大值"""
    value = ast.literal_eval(x)
    for i in range(len(value), 0, -1):
        if value[i-1] != "-":
            return value[i-1]

data_["攻击范围_max"] = data_["攻击范围"].apply(get_attack_range_value)
data_.drop(columns=["攻击范围"], inplace=True)
data_.tail()
```

```python
# 将 再部署时间 和 攻击间隔 的值中的s去掉，转换为分别转换为int和float
data_["再部署时间"] = data_["再部署时间"].apply(lambda x: int(x[:-1]))
data_["攻击间隔"] = data_["攻击间隔"].apply(lambda x: float(x[:-1]))

# 再将这两列的名字title后加上 (s}
data_.rename(columns={"再部署时间": "再部署时间(s}", "攻击间隔": "攻击间隔(s}"}, inplace=True)

data_.head()
```

```python
# 修正阻挡数
def get_min_max_block_value(x):
    """获取攻击范围的最大值"""
    if "→" in x:
        return int(x.split("→")[0]), int(x.split("→")[-1])
    else:
        return int(x), int(x)

data_["阻挡数_min"] = data_["阻挡数"].apply(lambda x: get_min_max_block_value(x)[0])
data_["阻挡数_max"] = data_["阻挡数"].apply(lambda x: get_min_max_block_value(x)[1])
data_.drop("阻挡数", axis=1, inplace=True)
```

```python
# 调整顺序
data_ = change_column(data_, '阻挡数_min', 5)
data_ = change_column(data_, '阻挡数_max', 6)
data_ = change_column(data_, '攻击范围_max', 7)
data_ = change_column(data_, '生命上限_trust', 11)
data_ = change_column(data_, '攻击_trust', 13)
data_ = change_column(data_, '防御_trust', 15)
data_ = change_column(data_, '法术抗性_trust', 17)
data_ = change_column(data_, '职业', -1)
data_.head()
```

```python
# 将除了"干员名称", "攻击间隔(s)", "职业"之外的数据的类型转换为int

except_list = ["干员名称", "攻击间隔(s)", "职业"]
for i in data_.columns:
    if i not in except_list:
        data_[i] = data_[i].astype(int)
```
