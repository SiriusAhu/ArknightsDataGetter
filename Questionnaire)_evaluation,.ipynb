# 测试表评估

```python
import pandas as pd
import os

path_a = 'data/answer.xlsx'
path_q = 'questionnaires'
paths_q = [q for q in os.listdir(path_q) if q.endswith('.xlsx')]

answer = pd.read_excel(path_a)
answer.head()
```

```python
qs = [pd.read_excel(os.path.join(path_q, path)) for path in paths_q]
qs[0].head()
```

```python
def accuracy_score(a):
    count = 0.
    all_count = len(a)
    false_df = pd.DataFrame()
    names = []
    x = []
    y = []
    for i in range(all_count):
        if a.iloc[i, 1] in a.iloc[i, 2]:
            count += 1
        else:
            names.append(a.iloc[i, 0])
            x.append(a.iloc[i, 1])
            y.append(a.iloc[i, 2])
    false_df['干员名称'] = names
    false_df['真值'] = x
    false_df['预测值'] = y
    return count / all_count, false_df

def acc_false_list_get(a, qs):
    accs = []
    false_dfs = []
    for q in qs:
        merge_columns = q.columns.tolist()[1:]
        merged_df = pd.merge(a, q, on=merge_columns)
        df = merged_df.iloc[:, [0, -2, -1]]
        acc, false_df = accuracy_score(df)
        accs.append(acc)
        false_dfs.append(false_df)
    return accs, false_dfs
```

```python
accs, false_dfs = acc_false_list_get(answer, qs)
```
